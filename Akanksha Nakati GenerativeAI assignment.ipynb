# What is Generative AI

Generative AI refers to a type of artificial intelligence technology that can generate new content, ranging from text and images to music and videos, after learning from a large dataset of existing examples. This technology relies on advanced machine learning models, particularly types like Generative Adversarial Networks (GANs), Variational Autoencoders (VAEs), and transformer-based models like GPT (Generative Pre-trained Transformer) and BERT (Bidirectional Encoder Representations from Transformers).

Here are some key points about Generative AI:

Learning and Generating: Generative AI models are trained on large amounts of data, learning patterns, styles, or features inherent in that data. After training, these models can generate new content that mimics the original data in a plausible way, often creating entirely new outputs that do not directly replicate the training examples.

Applications: Generative AI has a broad range of applications. In the creative arts, it can generate novel artworks, music compositions, and creative writing. In the business sector, it can be used for generating realistic product mock-ups, enhancing virtual reality environments, or automating content creation for marketing. It's also used in technology for tasks like data augmentation, which helps improve other AI models' performance by providing them with more diverse training data.

Ethical and Societal Implications: While generative AI holds great potential, it also raises ethical and societal concerns, such as the creation of deepfakes, copyright issues, and the authenticity of artistic creation. Addressing these challenges requires careful consideration of how these technologies are developed and deployed.

Continuous Evolution: The field of generative AI is rapidly evolving with ongoing research pushing the boundaries of what's possible. The introduction of more advanced models continues to expand the capabilities and efficiency of generative technologies.

## What is the generative AI technique being utilized?

![gpt2](/assignment/GPT%202.png)

The project employs the GPT-2 (Generative Pre-trained Transformer 2), a prominent model developed by OpenAI, known for its capabilities in generating coherent and contextually relevant text based on a given prompt. GPT-2 is built on the transformer architecture, which relies on self-attention mechanisms to process input data without the need for sequence-aligned recurrent processing, making it highly efficient for parallel computation and handling large datasets.

Model and Tokenizer Loading: The model (TFGPT2LMHeadModel or TFAutoModelForCausalLM) and tokenizer (AutoTokenizer) are loaded using libraries provided by Hugging Face, a platform known for hosting pre-trained models.
Text Tokenization: Input text is tokenized using the GPT-2 tokenizer. This process converts the text into a format that the model can understand, essentially breaking down the text into tokens that represent commonly occurring sequences of characters or words.
Text Generation: The model generates text based on the tokenized input. This is achieved by setting parameters such as max_length and num_return_sequences to control the length and number of generated sequences.
Output Processing: The generated sequences are then decoded back into human-readable text, demonstrating the model's ability to extend or enhance textual content based on learned patterns from the training data.
Applications Demonstrated in the Notebook:
using GPT-2 to generate synthetic text resembling movie overviews. This showcases the model's utility in creative content generation, where it can simulate various styles and forms of written content based on minimal input prompts.

## Theoretical Foundations of Generative AI

Introduction to Generative AI and Its Applications
Generative AI encompasses artificial intelligence technologies that enable machines to generate new content by learning from existing data. These models understand patterns, features, and relationships within the data and use this knowledge to produce new data instances that were not previously seen but are statistically similar to the original dataset. Common generative models include Generative Adversarial Networks (GANs), Variational Autoencoders (VAEs), and autoregressive models like GPT (Generative Pre-trained Transformer).

#### Applications of Generative AI:

**Art and Media:** Generating new artworks, music, and realistic images or videos.
**Text Generation:** Producing coherent and contextually relevant text for applications ranging from chatbots to creative writing.
**Synthetic Data Generation:** Creating data for training AI models where data may be scarce or sensitive.
**Drug Discovery:** Predicting molecular structures that could lead to new pharmaceuticals.

### The Relevance of Data Generation in Various Data Science Tasks
Data generation plays a crucial role in enhancing the robustness and efficiency of data science tasks:

**Model Robustness:** By generating diverse training examples, models can learn more generalized features of the data, reducing overfitting.
**Imbalanced Data:** Synthetic data generation helps address imbalances in datasets, particularly in classifications tasks where some classes are underrepresented.
**Privacy Concerns:** Synthetic data can mimic the statistical properties of real datasets while ensuring that the privacy of the underlying data subjects is maintained.

### Theoretical Underpinnings of GPT-2
GPT-2, as a chosen generative AI method, is based on the transformer architecture, which uses self-attention mechanisms to process input data. The model is pre-trained on a diverse corpus of text and then fine-tuned for specific tasks. This two-stage training process involves:

**Self-Attention:** The ability of the model to weigh the importance of different words in a sentence, regardless of their distance from each other.
Layered Architecture: Multiple layers of processing units allow GPT-2 to learn complex patterns in data, from basic syntactic structures to nuanced contextual relationships.
### How Generative AI Contributes to Solving Data-Related Problems
Generative AI significantly impacts various data-related problems:

**Data Scarcity:** AI models often require large amounts of data, which may not always be available. Generative models can produce additional data points.
**Anonymization:** In fields like healthcare, generative models can create anonymized datasets that are safe to use for research without compromising patient privacy.
**Innovation in Problem Solving:** Generative AI can propose solutions that may not be intuitive to human experts by exploring a broader solution space.
In conclusion, generative AI not only offers practical solutions to data insufficiency and privacy but also fosters innovation in AI applications, making it a cornerstone of modern AI research and application landscapes.

## Introduction to Data Generation

Data generation through generative AI involves creating new, synthetic data that is statistically similar to real-world data but does not directly replicate it. This process is essential in scenarios where data is scarce, sensitive, or needs augmentation to improve model training outcomes. Generative AI models capture the underlying distributions and patterns of the dataset they train on and use this learned information to produce new data instances.

The generation of synthetic data is increasingly vital across various domains:

**Training Data Augmentation:** Enhancing the volume and variety of data available for training machine learning models, especially in cases where gathering real-world data is challenging or expensive.
**Privacy Preservation:** Generating data that maintains the statistical properties of the original dataset while protecting individual privacy, crucial for complying with data protection regulations.
**Simulation and Testing:** Creating scenarios that might not be available in the collected data but are possible in real-world situations, helpful in fields like autonomous driving and financial modeling.
Principles Behind Data Generation
The key principles that underpin the data generation process in generative AI include:

**Diversity:** The generated data should cover the possible range of scenarios reflected in the original data, ensuring that models trained on this data can generalize well.
**Accuracy:** While the data is synthetic, its statistical properties must closely mirror those of the real data to be useful.
**Ethical Use:*8 Generated data must be created and used considering ethical guidelines, especially when personal data is involved.
**Data Generation Technique:** GPT-2
Description of the Chosen Generative AI Technique and Its Purpose in Data Generation
GPT-2 (Generative Pre-trained Transformer 2) is an advanced generative AI model designed primarily for text generation. Developed using the transformer architecture, it utilizes layers of self-attention mechanisms to understand and generate human-like text.

### Purpose in Data Generation:

**Text-Based Data Synthesis:** GPT-2 is proficient in generating realistic and coherent text passages, dialogues, or documents that can be used to augment text datasets. This is particularly useful in domains where textual data is necessary but hard to collect, such as dialogue generation for virtual assistants or generating realistic scenarios for training NLP models.
**Scenario Simulation:** It can simulate conversations or textual scenarios that might not be present in the original data, providing a richer dataset for training and testing NLP applications.
**Data Privacy:** By generating synthetic text, GPT-2 helps in creating datasets that contain no real user data, thus maintaining privacy and anonymity.

# Handling Missing Values

```python
import json

# Function to parse JSON-like strings into lists
def parse_json(data):
    try:
        return [item['name'] for item in json.loads(data.replace("'", "\""))]
    except:
        return []

# 1. Handling null values
# Fill missing values in 'overview' and 'genres' with an empty string or list
movies_data['overview'].fillna('', inplace=True)
movies_data['genres'] = movies_data['genres'].fillna('[]').apply(parse_json)

# Check the first few entries of 'genres' to confirm parsing and filling null values
movies_data.head(), movies_data.isnull().sum()
```

## Text Summarization

Let's move on to text summarization using a pre-trained NLP model. We'll use the Hugging Face transformers library to leverage a pre-trained GPT-2 model for generating concise summaries of movie descriptions from the "overview" column.

```python
%pip install --upgrade transformers tensorflow
```

First, we'll set up the model and tokenizer for GPT-2.

```python
# Ensure that the tokenizer and model are compatible
from transformers import TFGPT2LMHeadModel, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('gpt2')
model = TFGPT2LMHeadModel.from_pretrained('gpt2')
```

```python
# Define input text
input_text = "This is a sample overview of a movie."

# Tokenize input text
input_ids = tokenizer(input_text, return_tensors="tf").input_ids
```

```python
# Generate text
output_sequences = model.generate(
    input_ids=input_ids,
    max_length=50,  # adjust max_length as needed
    num_return_sequences=1,  # adjust num_return_sequences as needed
    pad_token_id=tokenizer.pad_token_id,
    eos_token_id=tokenizer.eos_token_id,
    no_repeat_ngram_size=2,
    early_stopping=True
)

# Decode and print generated text
generated_text = tokenizer.decode(output_sequences[0], skip_special_tokens=True)
print("Generated Text:", generated_text)
```
